```python
import pandas as pd
```

```python
# Save path to data set in a variable
data_file = "purchase_data.csv"
```

```python
# Use Pandas to read data
data_file_pd = pd.read_csv(data_file)
data_file_pd.head()
```

```python
# Total Number of Players
#Display the total number of players

total_players = data_file_pd["Gender"].count()
total_players = pd.DataFrame({"Total Players": [total_players]})
total_players
```

```python
### Purchasing Analysis (Total)

#Number of Unique Items
unique_items = len(data_file_pd["Item ID"].unique())

#Average Purchase Price
average_price = data_file_pd["Price"].mean()

#Total Number of Purchases
total_purchase = data_file_pd["Item Name"].count()

#Total Revenue
total_revenue = data_file_pd["Price"].sum()

total_summary = pd.DataFrame({"Unique Items": [unique_items],
                              "Average Price": [average_price],
                              "Number of Purchases": [total_purchase],
                              "Total Revenue": [total_revenue]})
                              
total_summary
```

```python
##Gender Demographics

#defining a variable 
total_gender = data_file_pd["Gender"].count()

#Percentage and Count of Male Players
male_count = data_file_pd[data_file_pd.Gender == 'Male'].count()
percentage_male = male_count / total_gender * 100

#Percentage and Count of Female Players
female_count = data_file_pd[data_file_pd.Gender == 'Female'].count()
percentage_female = female_count / total_gender * 100

#Percentage and Count of Other/Non-Disclosed
other_count = data_file_pd[data_file_pd.Gender == 'Other/Non-Disclosed'].count()
percentage_other = other_count / total_gender * 100

totals_summary = pd.DataFrame(index = ["Male", "Female", "Other/Non-Disclosed"], data = {"Percentage of Players": [percentage_male[5], percentage_female[5], percentage_other[5]],
                                        "Player Count": [male_count[5], female_count[5], percentage_other[5]]})
                                                       

#data_file_pd.iloc[1, data_file_pd.columns.get_loc("Percentage of Players")] = 'Male'
    
totals_summary 

#data_file = data_file_pd.set_index("Male", "Female", "Other")
```
